```python
# График заданной функции
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-0.5, 1.25, 1401)

plt.plot(x, x**3 - x**2)

plt.title('График функции y(x) = x**3 - x**2', fontsize=20, fontname='Times New Roman')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('tight')
plt.grid(True)
plt.show()
```

```python
# График производной - ищем экстремумы функции
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-0.5, 1, 1401)

plt.plot(x, 3*x**2 - 2*x)

plt.title('График функции y(x) = 3x**2 - 2x', fontsize=20, fontname='Times New Roman')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('tight')
plt.grid(True)
plt.show()
```
