```python
import math
def combinations (n, k):
  return int(factorial(n) / (factorial(k) * factorial(n-k)))
```

# 1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

В данном случае испытание повторяется многократно и исход каждого испытания независим от исходов других. Поэтому для решения задачи мы можем воспользоваться формулой Бернулли:

$P_n(k) = C_n^k \cdot p^k \cdot q^{n-k}$,

где $p = P(A)$ - вероятность появления события А в единичном испытании,

$q = 1 - p$ - вероятность противоположного события (событие А не наступило).

$p = 0.8$,

$q = 1 - 0.8 = 0.2$

$P_{100}(85) = C_{100}^{85} \cdot 0.8^{85} \cdot 0.2^{100-85} \approx 0.048$

```python
print(combinations(100,85) * 0.8**85 * 0.2**15)
```

Ответ: вероятность того, что стрелок из 100 выстрелов попадёт в цель ровно 85 раз, приблизительно равна 0,048.

# 2. Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?

В данном случае проводится достаточно большое количество испытанйий и при этом вероятность появления события $A$ в отдельном испытании мала. Следовательно, для решения задачи можно применить формулу Пуассона:

$P_m \approx \dfrac{\lambda^m}{m!}e^{-\lambda}$,

где $\lambda = n\cdot p$,

$n$ - количество испытаний,

$p$ - вероятность появления события $A$,

$m$ - количество появления событий в $n$ испытаниях.

$\lambda = 5000 \cdot 0.0004 = 2$

"Ни одна лампочка не перегорит в первый день" $\Rightarrow m = 0$.

$P_0 \approx \dfrac{2^0}{0!} \cdot 2.72^{-2} \approx 0.135$

"Перегорят ровно две лампочки" $\Rightarrow m=2$

$P_2 \approx \dfrac{2^2}{2!} \cdot 2.72^{-2} \approx 0.271$

```python
p0 = 2**0 / int(factorial(0)) * math.e**(-2)
print("P_0 = ", p0)
p2 = 2**2 / int(factorial(2)) * math.e**(-2)
print("P_2 = ", p2)
```

Ответ: вероятность того, что в первый день ни одна лампочка не перегорит, приблизительно равна 0,135.

Вероятность того, что в первый день перегорит ровно две лампочки, приблизительно равна 0,271.

# 4. В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что все мячи белые? Какова вероятность того, что ровно два мяча белые? Какова вероятность того, что хотя бы один мяч белый?

Вероятность - это отношение количества благоприятных исходов к общему количеству возможных:

$P(A) = \dfrac{m}{n}$, где

$m$ — число благоприятных исходов $A$;

$n$ — общее число возможных исходов.

a) Количество способов, которыми можно достать 2 белых мяча из первого ящика, равно:

$C_{7}^{2} = \dfrac{7!}{2! \cdot(7-2)!} = 21$

Количество способов, которыми можно достать 2 мяча из первого ящика, равно:

$C_{10}^{2} = \dfrac{10!}{2! \cdot(10-2)!} = 45$

Вероятность достать 2 белых мяча из первого ящика равна:

$P_1 = \dfrac{21}{45}$

Количество способов, которыми можно достать 2 белых мяча из второго ящика, равно:

$C_{9}^{2} = \dfrac{9!}{2! \cdot(9-2)!} = 36$

Количество способов, которыми можно достать 2 мяча из второго ящика, равно:

$C_{11}^{2} = \dfrac{11!}{2! \cdot(11-2)!} = 55$

Вероятность достать 2 белых мяча из второго ящика равна:

$P_2 = \dfrac{36}{55}$

Общая вероятность достать 2 белых мяча из первого ящика и 2 белых мяча из второго (т.е. все мячи белые) равна произведению соответствующих вероятностей:

$P = P_1 \cdot P_2 = \dfrac{21}{45} \cdot \dfrac{36}{55} = 0.30(54)$

б) Достать ровно 2 белых мяча из двух ящиков можно тремя способами: 2 белых мяча из первого (и два 2 чёрных мяча из второго), 2 белых мяча из второго (и 2 чёрных мяча из первого), по 1 белому мячу (и по 1 чёрному) из каждого ящика. Найдём вероятность каждого из случаев.

Вероятность вытащить 2 белых мяча из первого ящика и 2 чёрных мяча из второго равна:

$P_1 = \dfrac{C_7^2}{C_{10}^2} \cdot \dfrac{C_2^2}{C_{11}^2} = \dfrac{21}{45} \cdot \dfrac{1}{55} = \dfrac{21}{2475}$

Вероятность вытащить 2 белых мяча из второго ящика и 2 чёрных мяча из первого равна:

$P_2 = \dfrac{C_9^2}{C_{11}^2} \cdot \dfrac{C_3^2}{C_{10}^2} = \dfrac{36}{55} \cdot \dfrac{3}{45} = \dfrac{108}{2475}$

Вероятность вытащить один белый мяч и один чёрный из первого ящика равна:

$P_{3_1} = \dfrac{C_7^1 \cdot C_3^1}{C_{10}^2} = \dfrac{21}{45}$

Вероятность вытащить один белый мяч и один чёрный из второго ящика равна:

$P_{3_2} = \dfrac{C_9^1 \cdot C_2^1}{C_{11}^2} = \dfrac{18}{55}$

Вероятность достать по 1 мячу каждого цвета из каждого ящика равна:

$P_3 = P_{3_1} \cdot P_{3_2} = \dfrac{21}{45} \cdot \dfrac{18}{55} = \dfrac{378}{2475}$

Общая вероятность достать 2 белых мяча из двух ящиков равна сумме вероятностей каждого из рассмотреных вариантов:

$P = P_1 + P_2 + P_3 = \dfrac{21}{2475} + \dfrac{108}{2475} + \dfrac{378}{2475} = \dfrac{507}{2475} = 0.20(48)$


в) Вероятность того, что будет вытащен хотя бы один белый шар, равна разнице между 1 и вероятностью того, что вытащили все шары чёрного цвета.

Количество способов, которыми можно достать 2 чёрных мяча из первого ящика, равно:

$C_{3}^{2} = \dfrac{3!}{2! \cdot1!} = 3$

Количество способов, которыми можно достать 2 мяча из первого ящика, равно:

$C_{10}^{2} = \dfrac{10!}{2! \cdot(10-2)!} = 45$

Вероятность достать 2 чёрных мяча из первого ящика равна:

$P_1 = \dfrac{3}{45}$

Количество способов, которыми можно достать 2 чёрных мяча из второго ящика, равно:

$C_{2}^{2} = \dfrac{2!}{2! \cdot 2!} = 1$

Количество способов, которыми можно достать 2 мяча из второго ящика, равно:

$C_{11}^{2} = \dfrac{11!}{2! \cdot(11-2)!} = 55$

Вероятность достать 2 чёрных мяча из второго ящика равна:

$P_2 = \dfrac{1}{55}$

Общая вероятность достать 2 чёрных мяча из первого ящика и 2 чёрных мяча из второго (т.е. все мячи чёрные) равна произведению соответствующих вероятностей:

$P_ч = P_1 \cdot P_2 = \dfrac{3}{45} \cdot \dfrac{1}{55} = 0.00(12)$

Соответственно, вероятность того, что из ящика будет взят хотя бы один белый шар, равна:

$P_б = 1 - P_ч = 1 - 0.00(12) = 0,99(87)$
