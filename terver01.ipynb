# 1. Из колоды в 52 карты извлекаются случайным образом 4 карты.
# a) Найти вероятность того, что все карты – крести.
# б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.



а) Вероятность - это отношение количества благоприятных исходов к общему количеству возможных:

$P(A) = \dfrac{m}{n}$, где

$m$ — число благоприятных исходов $A$;

$n$ — общее число возможных исходов.

Всего в колоде 52 карты и 4 масти, следовательно, каждой масти в колоде 13 карт. Количество способов, которыми можно извлечь 4 карты крестовой масти из всех крестовых карт, равно:

$C_{13}^{4} = \dfrac{13!}{4! \cdot(13-4)!} = 715$

Количество способов, которыми можно извлечь 4 карты из всей колоды карт, равно:

$C_{52}^{4} = \dfrac{52!}{4! \cdot(52-4)!} = 270725$

Искомая вероятность равна:

$P = \dfrac{715}{270725} \approx 0,00264$, т.е. приблизительно четверть процента.

б) Вначале найдём вероятность того, что в вытащенных 4-х картах не окажется туза.
Количество способов, которыми можно извлечь 4 карты из 48 (без тузов), равно:

$C_{48}^{3} = \dfrac{48!}{3! \cdot(48-3)!} = 17296$

Количество способов, которыми можно извлечь 4 карты из всей колоды карт, равно:

$C_{52}^{4} = \dfrac{52!}{4! \cdot(52-4)!} = 270725$

Вероятность вытащить 4 карты, среди которых не окажется туза, равна:

$P_1 = \dfrac{17296}{270725}$

Искомая вероятность - это разность между 1 и вероятностью вытащить карты без хотя бы одного туза.

$P = 1 - \dfrac{17296}{270725} = \dfrac{253429}{270725} \approx 0,936$

# 2. На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

Количество способов, которыми можно выбрать 3 кнопки из 10 (число сочетаний), равно:

$C_{10}^{3} = \dfrac{10!}{3! \cdot(10-3)!} = 120$

Верная комбинация, которая откроет дверь, только одна.
Следовательно, вероятность того, что человек, не знающий код, откроет замок с первой попытки, равна:

$P = \dfrac{1}{120} \approx 0,0083$

# 3. В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

Количество способов, которыми можно извлечь 3 детали из 9 окрашенных, равно:

$C_{9}^{3} = \dfrac{9!}{3! \cdot 6!} = 84$

Количество способов, которыми можно извлечь 3 детали из общей кучи, равно:

$C_{15}^{3} = \dfrac{15!}{3! \cdot 12!} = 455$

Искомая вероятность равна отношению благоприятных исходов к общему их количеству:

$P = \dfrac{84}{455} \approx 0,1846$

# 4. В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

Очевидно, что количество способов, которыми можно купить 2 билета из 2 выигрышных, равно 1.

Количество способов, которыми можно купить 2 билета из 100, равно:

$C_{100}^{2} = \dfrac{100!}{2! \cdot 98!} = 4950$

Искомая вероятность равна отношению благоприятных исходов к общему их количеству:

$P = \dfrac{1}{4950} \approx 0,0002$
